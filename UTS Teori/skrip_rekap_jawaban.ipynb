```python
import os
import zipfile
import pandas as pd
```

```python
def get_answer(folderpath, fileoutput):
    agregat = {}
    # num_PG = []
    for i in range(60):
        agregat["PG_"+str(i+1)] = []
        # num_PG.append(i+1)
    for i in range(10):
        agregat["PAIRS_"+str(i+1)] = []
    for i in range(3):
        agregat["ESAI_"+str(i+1)] = []
    NPM = []
    for filename in os.listdir(folderpath):
        if filename.endswith(".zip"):
            zip_path = os.path.join(folderpath, filename)
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                for inner_file_name in zip_ref.namelist():
                    if inner_file_name.endswith(".xlsx"):
                        with zip_ref.open(inner_file_name) as xlsx_file:
                            try:
                                new_NPM = str(pd.read_excel(xlsx_file, sheet_name="Identitas").iloc[5, 3])
                                NPM.append(new_NPM)
                                PG = pd.read_excel(xlsx_file, sheet_name="PG")
                                PG = list(PG[~PG['Unnamed: 0'].isna()]['Unnamed: 5'])
                                for i in range(60):
                                    agregat["PG_"+str(i+1)].append(PG[i])
                                PAIRS = pd.read_excel(xlsx_file, sheet_name="Pairs")
                                PAIRS = list(PAIRS[~PAIRS['MENJODOHKAN'].isna()]["BOBOT: 10%"])
                                for i in range(10):
                                    agregat["PAIRS_"+str(i+1)].append(PAIRS[i])
                                ESAI = pd.read_excel(xlsx_file, sheet_name="Esai")
                                for i in range(3):
                                    ESAI = pd.read_excel(xlsx_file, sheet_name="Esai")
                                    jawaban = ESAI[(ESAI["Unnamed: 0"] == "Jawaban") | (ESAI["Unnamed: 0"] == "Jawaban:")].index
                                    esai_compile = []
                                    for idx, i in enumerate(jawaban):
                                        jaw = ""
                                        if i != jawaban[-1]:
                                            add = ESAI.iloc[i:jawaban[idx+1],1]
                                        else:
                                            add = ESAI.iloc[i:len(ESAI), 1]
                                        if str(add.iloc[0]) != 'nan':
                                            for row in add:
                                                if row and str(row) != "nan":
                                                    jaw += str(row)
                                        esai_compile.append(jaw)
                                    for i in range(6):
                                        agregat["ESAI_"+str(i+1)].append(esai_compile[i])
                            except:
                                print(f"Error processing {inner_file_name}")
    agregat['NPM'] = NPM
    OUTPUT = pd.DataFrame(agregat)
    OUTPUT.to_excel(fileoutput)
    return OUTPUT
```
